```python
import pandas as pd
import requests
import time
import os
import logging
# from scrapy import Selector
# from scrapy.crawler import CrawlerProcess
# import scrapy
import boto3
from dotenv import load_dotenv
from sqlalchemy import create_engine

# import numpy as np
# from fake_useragent import UserAgent


load_dotenv()
AWS_KEY = os.environ["AWS_ACCESS_KEY_ID"]
AWS_SECRET_KEY = os.environ["AWS_SECRET_ACCESS_KEY"]
BUCKET = os.environ["BUCKET"]
CONNECTION_STRING = os.environ["CONNECTION_STRING"]
```

```python
data_path = {
    "gps" : r"data_gps.csv",
    "weather" : r"data_weather.csv",
    "booking" : r"data_booking.csv"
}

session = boto3.Session(aws_access_key_id = AWS_KEY, 
                        aws_secret_access_key = AWS_SECRET_KEY, 
                        region_name = "eu-west-3")

s3 = session.resource("s3")
data_bucket = s3.Bucket(BUCKET)
```

```python
dl_data = []
for item in data_bucket.objects.all():
        if any(data in item.key for data in data_path.values()):
            data_bucket.download_file(item.key, f"data\dl_{item.key}")
```

```python
data_set = {}
for key, path in data_path.items():
    data_set[key] = pd.read_csv(f"dl_S3_{path}", index_col=0)
```

```python
pd.DataFrame().drop_duplicates().drop
```

```python
for key, df in data_set.items():
    df.dropna(inplace = True)
    df.drop_duplicates(inplace = True)

    data_set[key] = df
```

```python
engine = create_engine(CONNECTION_STRING)
for key, value in data_set.items():
    value.to_sql(
        name=key,       # Nom de la table
        con=engine,
        if_exists="replace",   # "replace", "append" ou "fail"
        index=False            # Ne pas sauvegarder l'index pandas
    )
```
